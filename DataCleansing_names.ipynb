```python
import pandas as pd
import glob

data = pd.DataFrame()
cols = ['state', 'gender', 'year', 'name', 'count']
df=[]
for f in glob.glob('*.txt'):
    
    df.append(pd.read_csv(f,names=cols))
```

```python
df=pd.concat(df,axis='rows')
df.to_csv("final_names_data.csv")
```

```python
df.sort_index(ascending=False,inplace=False)
```
